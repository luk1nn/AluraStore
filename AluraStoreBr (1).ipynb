### Importação dos dados



```python
import pandas as pd

import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
soma1 = loja['Preço'].sum()
soma2 = loja2['Preço'].sum()
soma3 = loja3['Preço'].sum()
soma4 = loja4['Preço'].sum()

```

```python
print(f"A soma da loja 1 e: {soma1}")
print(f"A soma da loja 2 e: {soma2}")
print(f"A soma da loja 3 e: {soma3}")
print(f"A soma da loja 4 e: {soma4}")
```

# 2. Vendas por Categoria


```python
venda_categoria1 = loja.groupby('Categoria do Produto')['Preço'].sum()
venda_categoria2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
venda_categoria3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
venda_categoria4 = loja4.groupby('Categoria do Produto')['Preço'].sum()
```

```python
print(f'Essas sao as vendas por categoria da loja 1: {venda_categoria1}')
print(f'Essas sao as vendas por categoria da loja 2: {venda_categoria2}')
print(f'Essas sao as vendas por categoria da loja 3: {venda_categoria3}')
print(f'Essas sao as vendas por categoria da loja 4: {venda_categoria4}')
```

# 3. Média de Avaliação das Lojas

```python
media1 = loja['Avaliação da compra'].mean()
media2 = loja2['Avaliação da compra'].mean()
media3 = loja3['Avaliação da compra'].mean()
media4 = loja4['Avaliação da compra'].mean()
```

```python
print(f'A Media da loja 1 eh: {media1}')
print(f'A Media da loja 2 eh: {media2}')
print(f'A Media da loja 3 eh: {media3}')
print(f'A Media da loja 4 eh: {media4}')
```

# 4. Produtos Mais e Menos Vendidos

```python
venda_produtos1 = loja.groupby('Produto')['Produto'].count()
venda_produtos2 = loja2.groupby('Produto')['Produto'].count()
venda_produtos3 = loja3.groupby('Produto')['Produto'].count()
venda_produtos4 = loja4.groupby('Produto')['Produto'].count()
```

```python
print(venda_produtos1.sort_values())
print(venda_produtos2.sort_values())
print(venda_produtos3.sort_values())
print(venda_produtos4.sort_values())




```

# 5. Frete Médio por Loja


```python
mediafrete1 = round(loja['Frete'].mean(),2)
mediafrete2 = round(loja2['Frete'].mean(),2)
mediafrete3 = round(loja3['Frete'].mean(),2)
mediafrete4 = round(loja4['Frete'].mean(),2)
```

```python
print(f'A Media da loja 1 eh: {mediafrete1}')
print(f'A Media da loja 2 eh: {mediafrete2}')
print(f'A Media da loja 3 eh: {mediafrete3}')
print(f'A Media da loja 4 eh: {mediafrete4}')
```

```python

labels = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
sizes = [soma1, soma2, soma3, soma4]
colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange']
explode = (0.05, 0.05, 0.05, 0.05)

fig, ax = plt.subplots()
ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    explode=explode,
    shadow=True
)

ax.set_title('Participação das Lojas nas Vendas Totais')
ax.axis('equal')

plt.tight_layout()
plt.show()
```

```python

labels = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
sizes = [media1, media2, media3, media4]
bar_labels = ['red', 'blue', 'black', 'orange']

fig, ax = plt.subplots()
bars = ax.bar(labels, sizes, label=bar_labels)

ax.bar_label(bars, padding=3)

ax.set_ylabel('Média de Vendas')
ax.set_title('Média de avaliação por Loja')
plt.tight_layout()

plt.show()
```

```python
# Grafico loja4 1
loja4['Data da Compra'] = pd.to_datetime(loja4['Data da Compra'], dayfirst=True, errors='coerce')


loja4['Ano'] = loja4['Data da Compra'].dt.year

loja4['Total da Venda'] = loja4['Preço']

vendas_por_ano = loja4.groupby('Ano')['Total da Venda'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(vendas_por_ano['Ano'], vendas_por_ano['Total da Venda'], marker='o', linestyle='-')
plt.title('Evolução das Vendas por Ano')
plt.xlabel('Ano')
plt.ylabel('Total Vendido (R$)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
